```python
#  Check available columns before using them
print("Available columns in df_resampled:", df_resampled.columns)

#  Ensure pollutant columns exist
pollutants = ["CO(GT)", "NO2(GT)", "O3(GT)"]
available_pollutants = [p for p in pollutants if p in df_resampled.columns]

#  If O3(GT) is missing, handle NaNs
if "O3(GT)" in df_resampled.columns:
    df_resampled["O3(GT)"].fillna(df["O3(GT)"].mean(), inplace=True)

#  1. Create Histogram for CO levels
if "CO(GT)" in available_pollutants:
    plt.figure(figsize=(8, 5))
    df_resampled["CO(GT)"].hist(bins=30, color='skyblue', edgecolor='black')
    plt.title("Distribution of CO Levels")
    plt.xlabel("CO (GT)")
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.savefig("CO_distribution.pdf")
    plt.show()
else:
    print("Error: 'CO(GT)' column is missing.")

#  2. Scatter Plot for NO2 vs. O3
if "NO2(GT)" in available_pollutants and "O3(GT)" in available_pollutants:
    plt.figure(figsize=(8, 5))
    plt.scatter(df_resampled["NO2(GT)"], df_resampled["O3(GT)"], alpha=0.5, color='purple')
    plt.title("Scatter Plot of NO2 vs. O3 Levels")
    plt.xlabel("NO2 (GT)")
    plt.ylabel("O3 (GT)")
    plt.grid(True)
    plt.savefig("NO2_O3_scatter.pdf")
    plt.show()
else:
    print("Error: 'NO2(GT)' or 'O3(GT)' column is missing.")

#  3. Box Plot of O3 Levels by Month
if "O3(GT)" in available_pollutants:
    df_resampled["Month"] = df_resampled.index.month  # Extract month
    plt.figure(figsize=(10, 6))
    df_resampled.boxplot(column="O3(GT)", by="Month")
    plt.title("Boxplot of O3 Levels by Month")
    plt.xlabel("Month")
    plt.ylabel("O3 (GT) Concentration")
    plt.grid(True)
    plt.savefig("O3_boxplot.pdf")
    plt.show()
else:
    print("Error: 'O3(GT)' column is missing.")

#  4. Heatmap of Correlation Between Pollutants
if len(available_pollutants) > 1:  # Need at least 2 pollutants for correlation
    import seaborn as sns
    plt.figure(figsize=(8, 6))
    sns.heatmap(df_resampled[available_pollutants].corr(), annot=True, cmap="coolwarm", fmt=".2f")
    plt.title("Heatmap of Pollutant Correlation")
    plt.savefig("pollutant_heatmap.pdf")
    plt.show()
else:
    print("Error: Not enough pollutant columns for correlation heatmap.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Ensure required pollutants exist
pollutants = ["CO(GT)", "NO2(GT)", "O3(GT)"]
available_pollutants = [p for p in pollutants if p in df.columns]

if not available_pollutants:
    print("Error: None of the specified pollutants are available in the dataset.")
else:
    # 1 Rolling Average (7-day window) for CO levels
    if "CO(GT)" in df.columns:
        df["CO_7DayAvg"] = df["CO(GT)"].rolling(window=7).mean()
        plt.figure(figsize=(10, 5))
        plt.plot(df.index, df["CO(GT)"], label="CO Levels", alpha=0.5)
        plt.plot(df.index, df["CO_7DayAvg"], label="7-Day Rolling Avg", color='red')
        plt.title("7-Day Rolling Average of CO Levels")
        plt.xlabel("Date")
        plt.ylabel("CO (GT)")
        plt.legend()
        plt.grid(True)
        plt.savefig("CO_rolling_avg.pdf")
        plt.show()

    # 2 Seasonal Trends in Pollutant Levels
    df["Month"] = df.index.month
    plt.figure(figsize=(10, 5))
    for pollutant in available_pollutants:
        monthly_avg = df.groupby("Month")[pollutant].mean()
        plt.plot(monthly_avg.index, monthly_avg, label=pollutant)
    plt.title("Seasonal Trends in Pollutant Levels")
    plt.xlabel("Month")
    plt.ylabel("Average Pollutant Level")
    plt.xticks(range(1, 13), ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
    plt.legend()
    plt.grid(True)
    plt.savefig("seasonal_trends.pdf")
    plt.show()

    # 3 Group Data by Hour of the Day
    df["Hour"] = df.index.hour
    hourly_avg = df.groupby("Hour")[available_pollutants].mean()

    plt.figure(figsize=(10, 5))
    for pollutant in available_pollutants:
        plt.plot(hourly_avg.index, hourly_avg[pollutant], label=pollutant)
    plt.title("Average Pollutant Levels by Hour of the Day")
    plt.xlabel("Hour of the Day")
    plt.ylabel("Average Concentration")
    plt.xticks(range(0, 24))
    plt.legend()
    plt.grid(True)
    plt.savefig("hourly_trends.pdf")
    plt.show()
```
